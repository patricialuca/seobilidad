```python
# Creamos una variable de tipo diccionario con los valores de los que hemos partido
# donde iremos almacenando los datos de frecuencia de competidores que aparecen en los
# datos que hemos cargado
diccionario = {
    "bitiba.es": 1,
    "kiwoko.com": 1,
    "mascotaplanet.com": 1,
    "mascoteros.com": 1,
    "miscota.es": 1,
    "nuestroperro.es": 1,
    "petclic.es": 1,
    "piensoymascotas.com": 1,
    "tiendanimal.es": 1
}

# creamos una función a la que le pasaremos cada dataframe
def creaDiccionario (df):
  # recorremos por filas el dataframe
  for i in df.index:
    # para cada fila de la columna donde aparece la información del dominio de la competencia
    # si no está en el diccionario lo añadimos con valor uno por su primera aparición
    if df['Dominio de la competencia'][i] not in diccionario:
      diccionario[df['Dominio de la competencia'][i]] = 1
    # si ya está en el diccionario, entonces incrementamos en uno el valor
    else:
      diccionario[df['Dominio de la competencia'][i]] += 1

# a continuación pasamos cada uno de los dataframe por la función para que cree el diccionario
creaDiccionario(dfCompetidoresBitiba)
creaDiccionario(dfCompetidoresKiwoko)
creaDiccionario(dfCompetidoresMascotaPlanet)
creaDiccionario(dfCompetidoresMascoteros)
creaDiccionario(dfCompetidoresMiscota)
creaDiccionario(dfCompetidoresNuestroPerro)
creaDiccionario(dfCompetidoresPetclic)
creaDiccionario(dfCompetidoresPiensoyMascotas)
creaDiccionario(dfCompetidoresTiendanimal)
```

## 2.1. Palabras clave

```python
# Creamos una variable de tipo diccionario donde iremos guardando todas las palabras clave que hay entre todos los competidores
# y además iremos guardando el número de veces que se repite para localizar todos los competidores que compiten por la misma palabra clave
keywords = {}

# creamos una función a la que le pasaremos cada dataframe de cada dominio
def creaKeywords (dfk, n):
  # recorremos por filas el dataframe
  for i in dfk.index:
    # para cada fila de la columna donde aparece la información de la palabra clave
    # si no está en el diccionario lo añadimos con valor uno por su primera aparición y almacenamos también el volumen de búsquedas
    if dfk['Palabras clave'][i] not in keywords:
      keywords[dfk['Palabras clave'][i]] = {'Sitio Web': listaNombres[n],'Aparición': 0,'Volumen': dfk['Volumen'][i], 'Estimación Visitas': dfk['Est Visitas'][i], 'Posición': dfk['Posición'][i]}
    # si ya está en el diccionario, entonces incrementamos en uno el valor de la aparición
    # el volumen de búsquedas se deja como está pues va a ser el mismo para la misma keyword
    else:
      keywords[dfk['Palabras clave'][i]]['Aparición'] += 1

# a continuación pasamos cada uno de los dataframe por la función  creada creaKeywords junto con el indice de webs para que cree el diccionario completo de palabras clave
indice = 0
for lista in listaDatos:
  creaKeywords(lista, indice)
  indice += 1
```

```python
# Pasamos el diccionario que hemos creado a un dataframe
dfKeywords = pd.DataFrame([key for key in keywords.keys()], columns=['Palabras Clave'])
dfKeywords['Aparición'] = [value['Aparición'] for value in keywords.values()]
dfKeywords['Volumen'] = [value['Volumen'] for value in keywords.values()]
dfKeywords
```

```python
# ordenamos el dataframe creado por cantidad de veces que aparece una palabra clave en orden ascendente
# y mostramos los primeros diez resultados
dfKeywordsOrdenado = dfKeywords.sort_values(by='Aparición', ascending=False)
dfKeywordsOrdenado.head(10)
```

Vemos que hay varias palabras clave que se repiten entre los 16 de los 18 competidores que estamos analizando. Entre ellas la palabra clave **'tiendas de perros'** es la que más volumen de búsquedas mensuales tiene, asi que la elegimos para obtener el ranking de posiciones.

# 3. Ranking de competidores para la palabra clave elegida

```python
# Creamos una variable de tipo diccionario donde iremos almacenando todos los datos extraidos
ranking = {}
# Guardamos en una variable nuestra palabra clave objetivo
keyword = 'tiendas de perros'
# creamos una función a la que le pasaremos cada archivo de datos de cada uno de los competidores
def creaRanking (df, n):
  # recorremos por filas el dataframe
  for i in df.index:
    if df['Palabras clave'][i] == keyword:
      ranking[df['Posición'][i]] = {'Sitio Web': listaNombres[n], 'Palabras Clave': df['Palabras clave'][i], 'Volumen': df['Volumen'][i], 'Posición': df['Posición'][i]}

indice = 0
for lista in listaDatos:
  creaRanking(lista, indice)
  indice += 1

ranking
```

```python
# Pasamos el diccionario que hemos creado a un dataframe
dfRanking = pd.DataFrame([key for key in ranking.keys()], columns=['Sitio Web'])
dfRanking['Palabras Clave'] = [value['Palabras Clave'] for value in ranking.values()]
dfRanking['Sitio Web'] = [value['Sitio Web'] for value in ranking.values()]
dfRanking['Posición'] = [value['Posición'] for value in ranking.values()]
dfRanking
```

```python
# ordenamos el dataframe creado por ranking en orden ascendente
dfRankingOrdenado = dfRanking.sort_values(by='Posición', ascending=True)
dfRankingOrdenado.head(10)
```

Hemos obtenido el ranking de posiciones de nuestros 10 competidores para la palabra clave escogida y que analizaremos en profundidad

# 4. Autoridad del Dominio y Backlinks
Para la palabra clave elegida obtenemos los resultados de los 100 resultados de búsqueda, obtenemos de esta parte la autoridad de dominio y la cantidad de enlaces que apuntan a cada uno de los dominios o backlinks

```python
#Creamos una variable que almacena los diez competidores elegidos
listaCompetidores =['bitiba.es', 'kiwoko.com', 'mascoteros.com', 'maskokotas.com', 'miscota.es',
               'nuestroperro.es', 'petclic.es', 'tiendanimal.es', 'zoomalia.es', 'zooplus.es']
#Leemos los datos del archivo csv donde hemos recopilado las 100 SERPS
dfSERPS = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/serps/ubersuggest_SERP_analysis.csv', header = 'infer', encoding='utf8')
dfSERPS
```

```python
# Creamos una variable de tipo diccionario donde iremos almacenando todos los datos extraidos de los competidores a estudiar
serps = {}

# creamos una función a la que le pasaremos cada archivo de datos de cada uno de los competidores
def creaSerps (df, n):
  # recorremos por filas el dataframe
  for i in df.index:
    if df['Dominio'][i] == listaCompetidores[n]:
      serps[df['Dominio'][i]] = {'Sitio Web': listaCompetidores[n], 'Autoridad de Dominio': df['DA'][i], 'Enlaces': df['Enlaces'][i], 'URL': df['URL'][i], 'Visitas': df['Visitas'][i],}

indice = 0
for lista in listaCompetidores:
  creaSerps(dfSERPS, indice)
  indice += 1

serps
```

```python
# Pasamos el diccionario que hemos creado a un dataframe
dfSerps = pd.DataFrame([key for key in serps.keys()], columns=['Sitio Web'])
dfSerps['URL destino'] = [value['URL'] for value in serps.values()]
dfSerps['Enlaces'] = [value['Enlaces'] for value in serps.values()]
dfSerps['Autoridad de Dominio'] = [value['Autoridad de Dominio'] for value in serps.values()]
dfSerps['Visitas'] = [value['Visitas'] for value in serps.values()]
dfSerps
```

```python
# ordenamos el dataframe creado por estimación de visitas en orden descendente
dfSerpsOrdenado = dfSerps.sort_values(by='Visitas', ascending=False)
dfSerpsOrdenado.head(10)
```

A partir de esta tabla podemos obtener los datos de backlinks y la autoridad de dominio.
